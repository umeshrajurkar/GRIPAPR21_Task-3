### Step-4 Expolaray Data Analyst

```python
import seaborn as sns
figsize=(15,10)
sns.pairplot(df,hue='Sub-Category')
plt.show
```

#### From the above plot we can say that our data is not normal and it has some amount of outliers.

```python
corr = df.corr()
sbn.heatmap(corr,annot=True,cmap='Reds')
```

```python
sbn.countplot(x=df['Ship Mode'])
```

From above Standard Class is very high as compaire to the other class

```python
#Analysing Segment of people buying from the supermart.

df['Segment'].value_counts()
```

```python
standard=(df['Segment'].value_counts()/len(df['Segment'])*100).plot(kind='bar',color='b')
```

From the above graph we said that
the 50% of people belongs to consumer class And 
20-30% of the people belongs to Corporate and Home offices.

```python
plt.figure(figsize = (6,6))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

Here Office Supplies Is More than Technology And Furniture

```python
plt.figure(figsize=(19,10))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.title('SHIPMENT STATE WISE')

plt.show()
```

California has highest sales.and North Dakota, Wyoming and West Virginia has less sales. 

```python
df.groupby('Region')['Profit'].agg(sum).plot(kind='bar',color='r')
plt.ylabel('Profit')
plt.title('Profits in different regions')
plt.show()
```

```python
#Visualizing Profit and Discount relation Using line plot

plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'], df['Profit'], data=df)

```

Profit is positive when the discount is less than 10%
